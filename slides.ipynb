## Testando graficos

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-np.pi/2, np.pi/2, 100)
plt.figure(figsize=(10, 8))
plt.plot(t, np.tan(t))
plt.xlabel("x")
plt.ylabel("tan(x)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
